```python
from pathlib import Path

from lipid_gnn.lipid_graph import create_global_encoder, MartiniHeteroGraphBuilder
```

```python
data_dir_path = Path("/home/phillip/Goethe/Thesis/lipid-graph-nn/data/membrane_only/POPC100")
run_dir_path = data_dir_path / "run"
tpr_file_path = run_dir_path / "prun.tpr"
traj_file_path = run_dir_path / "prun.xtc"
    
```

```python
global_encoder = create_global_encoder(str(data_dir_path.parent))
```

```python
cutoff_angstroms = 11.0 
builder = MartiniHeteroGraphBuilder(
    tpr_file=tpr_file_path,
    trajectory_file=traj_file_path,
    spatial_cutoff=cutoff_angstroms,
    encoder=global_encoder
)
data = builder.process_frame(frame_idx=1)
```
